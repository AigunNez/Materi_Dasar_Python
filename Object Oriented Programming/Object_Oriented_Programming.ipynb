# Apa itu Object Oriented Programming (OOP) atau Pemrograman Berorientasi Objek di Python?
Pemrograman berorientasi objek (Object-Oriented Programming atau OOP) adalah paradigma pemrograman yang berfokus pada konsep "objek". Sebuah objek mengandung data dan kode:

- Data berupa properti (sering disebut atribut).
- Kode berupa metode (aksi yang dapat dilakukan oleh objek).

Dalam OOP, program dirancang dengan menggunakan kelas dan objek. Python mendukung berbagai pendekatan pemrograman, seperti pemrograman fungsional dan pemrograman modular. Salah satu pendekatan populer untuk menyelesaikan masalah pemrograman adalah dengan membuat objek.

- Object
- Method

# Class

Kelas (Class) adalah sekumpulan objek dengan atribut dan metode tertentu. Sebuah *class* adalah blueprint atau cetak biru untuk membuat objek. *Class* ini mendefinisikan atribut (data) dan metode (fungsi) yang akan dimiliki oleh objek yang dibuat berdasarkan kelas tersebut.

Komponen Utama dalam Kelas:
- Atribut: Variabel yang menyimpan data atau informasi tentang objek.
- Metode: Fungsi yang mendefinisikan aksi atau perilaku yang dapat dilakukan oleh objek.

Cara Membuat Kelas di Python
Untuk membuat kelas di Python, Anda akan menggunakan kata kunci `class`. Sebuah kelas adalah blueprint atau cetak biru yang mendefinisikan atribut dan metode yang dimiliki oleh objek yang dibuat dari kelas tersebut.

Berikut adalah beberapa konsep dasar yang perlu Kita pahami ketika bekerja dengan kelas di Python:
- Kata Kunci `class`: Digunakan untuk mendefinisikan sebuah kelas.
- Atribut `instance`: Atribut yang spesifik untuk setiap objek yang dibuat dari kelas.
- Atribut `class`: Atribut yang bersifat umum dan dimiliki oleh kelas itu sendiri, bukan objek-objek individual.
- Kata Kunci `self`: Digunakan untuk merujuk ke objek saat ini dalam kelas.
- Metode `__init__()`: Konstruktor yang dipanggil saat objek dibuat dan digunakan untuk menginisialisasi atribut instansi.

## Kata Kunci `class`
Dengan kata kunci `class`, kita dapat membuat sebuah kelas dalam Python seperti pada contoh di bawah ini:

```
class BookStore:
    pass
```

## Apa itu `self`?

Python menyediakan kata kunci `self` untuk merepresentasikan instansi dari sebuah kelas. `self` berfungsi sebagai penghubung untuk mengakses atribut atau metode yang dimiliki oleh kelas tersebut.

**Catatan:**
- Parameter `self` secara implisit adalah argumen pertama dari metode `__init__` (dan semua metode lainnya) dalam setiap kelas Python.
- `self` mengacu pada objek spesifik yang dibuat dari kelas tersebut.


```
class BookStore:
    def __init__(self, name, location):
        self.name = name
        self.location = location
```

## Apa itu `__init__` (konstruktor) di Python?
Metode `__init__` adalah metode unik yang diasosiasikan dengan setiap kelas Python. Python secara otomatis memanggil metode ini setiap kali sebuah objek dibuat dari kelas. Tujuannya adalah untuk menginisialisasi atribut kelas dengan nilai-nilai yang diberikan oleh pengguna.

Metode ini juga dikenal sebagai constructor dalam pemrograman berorientasi objek.

````python
class BookStore:
    def __init__(self):
        print("__init__() Konstruktor telah dipanggil...")

    def beli(self):
        print("Anda membeli buku")

B1 = BookStore()
````

## Atribut Instance
Atribut instance adalah atribut yang spesifik untuk setiap objek. Atribut ini biasanya didefinisikan sebagai parameter dalam metode `__init__`. Setiap objek dapat memiliki nilai atribut yang berbeda.

````python
class BookStore:
    def __init__(self, attrib1, attrib2):
        self.attrib1 = attrib1
        self.attrib2 = attrib2

# Membuat objek dengan atribut yang berbeda
store1 = BookStore("Knowledge Hub", "Jakarta")
store2 = BookStore("Book Haven", "Surabaya")

print(f"Store1: {store1.attrib1}, Location: {store1.attrib2}")
print(f"Store2: {store2.attrib1}, Location: {store2.attrib2}")
````

## Atribut Kelas (Class Attributes)
Berbeda dengan atribut instance, yang hanya terlihat pada level objek, atribut kelas bersifat global untuk semua objek yang dibuat dari kelas tersebut. Nilainya akan sama untuk semua instance kelas tersebut.

````python
class BookStore:
    instances = 0  # Atribut kelas

    def __init__(self, attrib1, attrib2):
        self.attrib1 = attrib1  # Atribut instance
        self.attrib2 = attrib2
        BookStore.instances += 1  # Mengupdate atribut kelas

# Membuat beberapa objek dari kelas
b1 = BookStore("Knowledge Hub", "Jakarta")
b2 = BookStore("Book Haven", "Surabaya")

# Mengakses atribut kelas
print("BookStore.instances:", BookStore.instances)
````

### Perbedaan Atribut Kelas vs Atribut Instance

|Atribut Kelas|Atribut Instance|
|---|---|
|Bersifat global untuk semua objek.|Khusus untuk objek tertentu.|
|Diakses menggunakan nama kelas|Diakses menggunakan self.|
|Nilainya sama untuk semua objek.|Nilainya unik untuk setiap objek.|

````python
class BookStore:
    instances = 0  # Atribut kelas

    def __init__(self, name, location):
        self.name = name  # Atribut instance
        self.location = location
        BookStore.instances += 1

# Membuat objek
store1 = BookStore("Knowledge Hub", "Jakarta")
store2 = BookStore("Book Haven", "Surabaya")

# Mengakses atribut kelas dan instance
print(f"Total instances: {BookStore.instances}")
print(f"Store1 Name: {store1.name}, Location: {store1.location}")
print(f"Store2 Name: {store2.name}, Location: {store2.location}")
````

## Atribut Kelas dan Atribut Instance

Sekarang kita mencoba membuat kelas Dog dengan satu atribut kelas species dan dua atribut instance yaitu name dan age.

````python
class Dog:
    species = "Canis familiaris"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def description(self):
        return f"{self.name} is {self.age} years old"

    def speak(self, sound):
        return f"{self.name} says {sound}"

MILO = Dog("MILO", 4)
OLLIE = Dog("OLLIE", 3)
````

Pada kode di atas, kita membuat dua instance baru dari kelas Dog:

- Milo, anjing berusia 4 tahun
- Ollie, anjing berusia 3 tahun

## Penjelasan Mengenai Metode `__init__`
Metode `__init__` dalam kelas Dog memiliki tiga parameter (self, name, dan age), namun hanya dua argumen yang diberikan ketika objek diinstansiasi. Mengapa demikian?

Ketika Anda membuat objek baru dari kelas Dog, Python secara otomatis membuat instance baru dan memberikan instance tersebut pada parameter pertama (self) dari metode `__init__()`. Oleh karena itu, Anda hanya perlu memberikan argumen name dan age, karena `self` sudah terikat secara otomatis ke objek yang sedang dibuat.

## Metode Instance

Metode instance adalah fungsi yang didefinisikan di dalam kelas dan hanya dapat dipanggil melalui instance dari kelas tersebut. Sama seperti `__init__()`, parameter pertama dari metode instance selalu `self`.

````python
class Dog:
    species = "Canis familiaris"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Metode instance
    def description(self):
        return f"{self.name} is {self.age} years old"

    def speak(self, sound):
        return f"{self.name} says {sound}"
````

## Menjalankan Metode Instance
Setelah Anda membuat instance objek `Dog`, Anda dapat memanggil metode instance menggunakan dot notation.

````python
miles = Dog("Miles", 4)

# Memanggil metode description
print(miles.description())  # Output: Miles is 4 years old

# Memanggil metode speak dengan suara 'Woof Woof'
print(miles.speak("Woof Woof"))  # Output: Miles says Woof Woof

# Memanggil metode speak dengan suara 'Bow Wow'
print(miles.speak("Bow Wow"))  # Output: Miles says Bow Wow
````

## Penggunaan Metode yang Baik
Dalam kelas `Dog`, metode `description()` mengembalikan informasi tentang objek Dog yang lebih berguna. Saat menulis kelas Anda sendiri, adalah ide yang baik untuk membuat metode yang mengembalikan string yang berisi informasi berguna tentang instance kelas. Namun, `description()` bukanlah cara yang paling Pythonic untuk melakukannya. Anda bisa menggunakan` __str__()` untuk tujuan serupa.

````python
class Dog:
    species = "Canis familiaris"

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name} is {self.age} years old"
````

Dengan menambahkan `__str__()`, Anda dapat menggunakan metode ini untuk mencetak objek secara langsung tanpa perlu memanggil metode secara eksplisit.

````python
miles = Dog("Miles", 4)
print(miles)  # Output: Miles is 4 years old
````

## Membuat Objek di Python

Kita telah melihat bahwa objek kelas dapat digunakan untuk mengakses atribut-atribut yang berbeda. Objek kelas juga dapat digunakan untuk membuat instance objek baru dari kelas tersebut (instansiasi objek). Prosedur untuk membuat objek baru serupa dengan pemanggilan fungsi.

```
harry = Person()
```

Kode di atas akan membuat instance objek baru dengan nama harry. Setelah objek dibuat, kita dapat mengakses atribut objek menggunakan prefix nama objek.

Atribut objek bisa berupa data atau metode. Metode objek adalah fungsi-fungsi yang terkait dengan kelas tersebut.

````python
class Person:
    "This is a person class"
    age = 30

    def greet(self):
        print("Hi")

# Membuat objek baru dari kelas Person
Deepali = Person()

# Mengakses metode greet dari kelas Person
print(Person.greet)  # Output: <function Person.greet at 0x0000015E7B226700>

# Memanggil metode greet milik objek Deepali
Deepali.greet()  # Output: Hi
````

### Metode Objek
Metode objek adalah fungsi yang terkait dengan objek dari kelas tersebut. Dalam contoh ini, `greet()` adalah metode yang berfungsi untuk mencetak Hi. Setiap objek yang dibuat dari kelas Person memiliki metode `greet()` yang dapat dipanggil.

```
<function Person.greet at 0x0000015E7B226700>
```
Ini menunjukkan alamat memori tempat fungsi greet disimpan, yang merujuk ke fungsi yang sama pada setiap objek yang dibuat.

## Pewarisan (Inheritance) dalam Pemrograman Berorientasi Objek (OOP)

Pewarisan adalah konsep inti dalam Object-Oriented Programming (OOP) yang memungkinkan sebuah kelas baru (child class atau kelas turunan) untuk mewarisi atribut dan metode dari kelas yang sudah ada sebelumnya (parent class atau kelas induk). Dengan pewarisan, Anda dapat menggunakan kembali kode yang sudah ada, menambah fungsionalitas baru, tanpa harus mengubah kode asli.

### Tujuan Pewarisan

1. **Menggunakan Kembali Kode (Code Reuse)**:
    Atribut dan metode yang didefinisikan di kelas induk dapat digunakan oleh kelas turunan, sehingga mengurangi pengulangan kode.
2. **Memperluas Fungsionalitas**:
    Kelas turunan dapat menambahkan fitur baru atau mengubah perilaku yang sudah ada tanpa memengaruhi kelas induk.
3. **Analogi Kehidupan Nyata**:
    Sama seperti anak mewarisi sifat atau properti dari orang tuanya, kelas turunan juga dapat mewarisi properti dari kelas induknya.

### Cara Mengimplementasikan Pewarisan di Python
Di Python, pewarisan diimplementasikan dengan cara menulis nama kelas induk di dalam tanda kurung setelah nama kelas turunan.
```
class KelasInduk:
    # Atribut dan metode kelas induk
    pass

class KelasTurunan(KelasInduk):
    # Atribut dan metode tambahan untuk kelas turunan
    pass

```

### Contoh Pewarisan


````python
# Kelas induk
class Hewan:
    def __init__(self, nama):
        self.nama = nama

    def berbicara(self):
        return f"{self.nama} membuat suara"

# Kelas turunan 1
class Anjing(Hewan):
    def berbicara(self):
        return f"{self.nama} menggonggong"

# Kelas turunan 2
class Kucing(Hewan):
    def berbicara(self):
        return f"{self.nama} mengeong"

# Membuat objek
anjing = Anjing("Buddy")
kucing = Kucing("Kitty")

print(anjing.berbicara())  # Output: Buddy menggonggong
print(kucing.berbicara())  # Output: Kitty mengeong
````

### Konsep Utama dalam Pewarisan Python
#### Method Overriding:
Kelas turunan dapat mengganti metode dari kelas induk untuk menyesuaikan kebutuhan.


````python
class Anjing(Hewan):
    def berbicara(self):
        return f"{self.nama} menggonggong dengan keras"
````

#### Fungsi super():
Digunakan untuk memanggil metode dari kelas induk.

````python
class Anjing(Hewan):
    def berbicara(self):
        return super().berbicara() + " dan menggonggong"

anjing = Anjing("Buddy")
print(anjing.berbicara())  # Output: Buddy menggonggong dengan kera
````

### Pewarisan Ganda (Multiple Inheritance):

Sebuah kelas dapat mewarisi lebih dari satu kelas induk.

````python
class Mamalia:
    pass

class Karnivora:
    pass

class Singa(Mamalia, Karnivora):
    pass
````

### Fungsi isinstance() dan issubclass():
- `isinstance(objek, Kelas)`: Mengecek apakah suatu objek merupakan instance dari kelas tertentu.
- `issubclass(KelasTurunan, KelasInduk)`: Mengecek apakah suatu kelas adalah turunan dari kelas lain.

````python
print(isinstance(anjing, Hewan))  # Output: True
print(issubclass(Anjing, Hewan)) # Output: True
````

### Keuntungan Pewarisan
1. **Pengurangan Duplikasi Kode**:
    Tidak perlu menulis ulang atribut dan metode yang sama di kelas turunan.2.
2. **Struktur Kode yang Lebih Baik**:
    Dengan pewarisan, hubungan hierarki antara kelas dapat dibentuk, seperti Hewan -> Anjing.
3. **Kemudahan Ekstensi**:
    Fitur baru dapat ditambahkan ke kelas turunan tanpa mengubah kelas induk.
